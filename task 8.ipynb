```python
import matplotlib.pyplot as plt

# Data: Days of the week and corresponding temperatures
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [22, 24, 21, 20, 23, 25, 26]

# Create the plot
plt.figure(figsize=(8, 5))  # Optional: Set figure size
plt.plot(days, temperatures, marker='o', linestyle='-', color='b', label='Temperature (°C)')

# Add labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.title('Weekly Temperature Variation')
plt.grid(True, linestyle='--', alpha=0.6)  # Optional: Add grid
plt.legend()  # Optional: Show legend

# Display the plot
plt.show()
```
